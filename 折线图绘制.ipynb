```python
import matplotlib.pyplot as plt
import numpy as np

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 准备数据
years = ['2022年', '2023年', '2024年']
y1 = [1337.72, 1114.59, 1127.73]  
y2 = [878.85, 618.74, 603.26]
y3 = [60.55, 48.86, 101.38]

# 创建双纵轴图
fig, ax1 = plt.subplots(figsize=(10, 6))

# 左轴：销售额
color1 = '#FF6B6B'
ax1.plot(years, y1, color=color1, marker='o', linewidth=3, 
         markersize=8, label='政府(市本级)总预算支出')
ax1.plot(years, y2, color=color1, marker='s', linewidth=3, 
         markersize=8, label='基础设施建设投资')
ax1.set_xlabel('年份', fontsize=12)
ax1.set_ylabel('政府(市本级)总预算支出/基础设施建设投资(亿元)', color=color1, fontsize=12)
ax1.tick_params(axis='y', labelcolor=color1)

# 右轴：利润率
ax2 = ax1.twinx()
color2 = '#4ECDC4'
ax2.plot(years, y3, color=color2, marker='^', linewidth=3, 
         markersize=8, label='交通设施建设投资')
ax2.set_ylabel('交通设施建设投资(亿元)', color=color2, fontsize=12)
ax2.tick_params(axis='y', labelcolor=color2)

# 添加图例
ax1.legend(loc='upper center', bbox_to_anchor=(0.38, 0.95))
ax2.legend(loc='upper center', bbox_to_anchor=(0.6, 0.95))


plt.title('厦门市2022-2024年财政投入趋势', fontsize=14, pad=20)
plt.tight_layout()
plt.show()
```
